**(2)** $f(x)=3x^2+3x-6$. Найдите значения функции при $x=0.5$ и $x=7$. <br>
**(3)** $f(x)=2x^4+x^3-6x^2+7x+1$. Найдите значения функции при $x=3$ и $x=–5$. <br>
**(4)** $f(x)=(x^2)^x$. Найдите значения функции при $x=0$, $х=10$. <br>


```
import sympy as  sp
from sympy.plotting import plot
```

```
x = sp.Symbol('x')
```

```
funcs = {
    1: [(-3 * x + 5), 0.5, 7],
    2: [(3 * x**2 + 3*x - 6), 3, -5],
    3: [(2*x**4 + x**3 -6*x**2 + 7*x + 1), 3, -5],
    4: [((x**2)**x), 0, 10]
}
```

```
for i in funcs:
  plot(funcs[i][0])
```

```
for i in funcs:
  for z in funcs[i][1:]:
    print(f"функция {funcs[i][0]} при x ={z} {funcs[i][0].subs(x,z)}")
```
